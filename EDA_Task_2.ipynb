```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".", "[.]")
```

```python
#find-numbers-with-even-number-of-digits
```

```python
from typing import List

class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        return sum(len(str(n)) % 2 == 0 for n in nums)
```

```python
#Number-of-good-pairs
```

```python
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        freq = {}
        good_pairs = 0
        
        for x in nums:
            good_pairs += freq.get(x, 0)   # each previous x forms a pair
            freq[x] = freq.get(x, 0) + 1
        
        return good_pairs
```

```python
#How-many-numbers-are-smaller-than-the-current-number
```

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        # sorted list of nums
        sorted_nums = sorted(nums)
        
        # map each number to its first index in sorted order
        rank = {}
        for i, val in enumerate(sorted_nums):
            if val not in rank:      # first occurrence = count of smaller numbers
                rank[val] = i
        
        # build answer
        return [rank[x] for x in nums]
        
```

```python
#Subtract-the-product-and-sum-of-digits-of-an-integer
```

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1
        total_sum = 0
        
        while n > 0:
            digit = n % 10
            product *= digit
            total_sum += digit
            n //= 10
        
        return product - total_sum
```

```python
#xor-operation-in-an-array
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0
        for i in range(n):
            result ^= (start + 2 * i)
        return result
```
